## Part 3: Write a query function (10 pts)

Write a function called `query_climate()` which accepts five arguments:

`df`, the file name for the dataframe

`country`, a string giving the name of a country for which data should be returned.

`year_begin` and `year_end`, two integers giving the earliest and latest years for which should be returned.

`month`, an integer giving the month of the year for which should be returned.

The return value of `query_climate()` is a Pandas dataframe of temperature readings for the specified country, in the specified date range, in the specified month of the year. This dataframe should have the following columns, **in this order**:

`NAME`: The station name.

`LATITUDE`: The latitude of the station.

`LONGITUDE`: The longitude of the station.

`Country`: The name of the country in which the station is located.

`Year`: The year in which the reading was taken.

`Month`: The month in which the reading was taken.

`Temp`: The temperature at the specified station during the specified year and month. 


**Grading is based on unseen test examples. A test example is given below.**

```python
def query_climate(df, country, year_begin, year_end, month):
    """Take a dataframe and extract all relevant temperature data based on a given country, month, 
       beginning and ending year.
       Args:
       - df
       - country
       - year_begin
       - year_end
       - month
       
       Returns:
       A dataframe containing the temperature data for a given country on a given month for 
       a specified length of years. It is returned with the column order: Station Name, Latitude, 
       Longitude, Country, Year, Month, Temperature.
     """
    if df["ID"].iloc[0]:
          df = df.drop(["ID"], axis = 1) #drop ID column
            
    df["Month"] = [month for i in range(len(df))]
    df["Temperature (C)"] = df[month]
    #rename the columns containing month and temperature data
    
    df = df[df["Country"] == country] #Get only rows with the right Country
    df = df[(df["Year"] >= year_begin) & (df["Year"] <= year_end)] #Get only rows with the right years
    df = df[df["Month"] == month] #Get only rows with the right month
    
    order = ["Station Name", "LATITUDE", "LONGITUDE", "Country", "Year", "Month", "Temperature (C)"]
    
    return df.sort_values(["Station Name",  "Year"]).loc[:, order]
    #return the df sorted by station name and year, in the specified order
```

```python
china = month_df[month_df["Country"] == "China"].copy()
        
china["Date"] = china["Year"].astype(str) + "-" + china["Month"].astype(str)
china["Date"] = pd.to_datetime(china["Date"])

china_max = china.groupby(["Date"])[["Temperature (C)"]].max()
china_max = china_max.reset_index() 

fig, ax = plt.subplots(figsize=(22, 3))
sns.lineplot(data = china_max,
             x = "Date",
             y = "Temperature (C)",
             ax = ax,
             color = "green"
            )

ax.set_title("Max Temperature in China over Time")

china_avg = china.groupby(["Date"])[["Temperature (C)"]].mean()
china_avg = china_avg.reset_index() 

fig, ax1 = plt.subplots(figsize=(22, 3))
sns.lineplot(data = china_avg,
             x = "Date",
             y = "Temperature (C)",
             ax = ax1,
             color = "orange"
            )

ax1.set_title("Average Temperature in China over Time")
```

For this line plot I sliced out a new dataframe containing only the data for China in order to explore the trend of its average temperature that we observed in the previous plot. I first used the 'Year' and 'Month' columns to combine them into a new column, 'Date'.
- For the top plot I took the max temperature recorded across all stations for each date (Year-Month) and plotted it over time.
- For the bottom plot I took the average temperature between all stations for each date (Year-Month) and plotted it over time.

Focusing in on China from the list of countries, we can see that although the average recorded temperature has gone down in more recent years, the max temperature shows an increase more in line of what we would expect from the effect of global warming and from what we see in the average temperature over time for other countries.

```python
china_lat = china.groupby(["LATITUDE"])[["Temperature (C)"]].mean()
china_lat= china_lat.reset_index() 

fig, ax = plt.subplots(1, figsize = (40,5))
sns.barplot(data = china_lat,
            
            x = "LATITUDE",
            y = "Temperature (C)",
            ax = ax,
            hue = "LATITUDE",
            legend = False,
            palette = "vlag_r",
           )

ax.tick_params(axis = 'x', labelrotation = 90, pad = 3, width = 0.5, labelsize = "small")
```

I wanted to explore the effect of latitude on the average temperatures across China. I used groupby() to find the mean temperature for each unique latitude value in the dataframe. I used this new subset of data to plot latitude vs. temperature and use a specific color palette to better visualize hotter and colder temperatures.

From the plot we observe that as we move towards the north pole and away from the equator as we go to the right on the x-axis, the average temperatures recorded at each station trends to colder climates. This is intuitive as the climate should get less tropical as we move away from the equator which results in the trend displayed across the country.
